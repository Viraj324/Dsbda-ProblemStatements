```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
col_names = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg',
             'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']
```

```python
c = pd.read_csv("Cleavland.csv", names=col_names, na_values="?", sep=",")
h = pd.read_csv("hung.csv", names=col_names, na_values="?", sep=",")
s = pd.read_csv("Switzerland.csv", names=col_names, na_values="?", sep=",")
```

```python
df = pd.concat([c,s,h])
```

```python
def draw_box(df, col):
    sns.boxplot(df[col])
    plt.title(f"Box plot {col}")  # Use f-string to include variable value in the title

for col in df.columns:
    draw_box(df, col)  # Pass each column name to the draw_box function
    plt.show()  # Display each plot separately
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df["age"],bins=20,kde=True)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Histogram for age")
```

```python
df.reset_index(drop=True,inplace=True)
sns.lineplot(x="age",y="trestbps",data=df)
```

```python
plt.figure(figsize=(10,6))
sns.stripplot(x='cp',y='age',data=df)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x="sex",y="chol",data = df)
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x="age",y="cp",hue="chol",data=df,s=100)
```
